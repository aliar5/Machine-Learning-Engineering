```python
# Tiny ImageNet Classification with ResNet-50

import os
import shutil
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
from torchvision import datasets, transforms, models
from torchvision.utils import make_grid
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report, top_k_accuracy_score
import seaborn as sns
from tqdm import tqdm
import copy

# Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Download and Prepare Tiny ImageNet
if not os.path.exists("tiny-imagenet-200"):
    !wget http://cs231n.stanford.edu/tiny-imagenet-200.zip
    !unzip -q tiny-imagenet-200.zip

val_dir = "tiny-imagenet-200/val"
val_images_dir = os.path.join(val_dir, "images")
val_annotations_file = os.path.join(val_dir, "val_annotations.txt")

if os.path.exists(val_images_dir):
    with open(val_annotations_file, 'r') as f:
        for line in f.readlines():
            img_name, class_name = line.split('\t')[:2]
            class_dir = os.path.join(val_dir, class_name)
            os.makedirs(class_dir, exist_ok=True)
            shutil.move(os.path.join(val_images_dir, img_name), os.path.join(class_dir, img_name))
    shutil.rmtree(val_images_dir)

# Transforms
transform_train = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(10),
    transforms.ToTensor(),
    transforms.Normalize((0.44, 0.44, 0.44), (0.22, 0.22, 0.22))
])

transform_val = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.44, 0.44, 0.44), (0.22, 0.22, 0.22))
])

# Data Loaders
data_dir = "/content/tiny-imagenet-200"
train_dataset = datasets.ImageFolder(os.path.join(data_dir, 'train'), transform=transform_train)
val_dataset = datasets.ImageFolder(os.path.join(data_dir, 'val'), transform=transform_val)

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True, num_workers=8)
val_loader = DataLoader(val_dataset, batch_size=64, shuffle=False, num_workers=8)
class_names = train_dataset.classes

# Custom ResNet50 with dropout
class CustomResNet50(nn.Module):
    def __init__(self):
        super(CustomResNet50, self).__init__()
        self.base = models.resnet50(pretrained=True)
        for param in self.base.parameters():
            param.requires_grad = False
        for layer in [self.base.layer2, self.base.layer3, self.base.layer4]:
            for param in layer.parameters():
                param.requires_grad = True
        self.base.fc = nn.Sequential(
            nn.Dropout(0.3),
            nn.Linear(self.base.fc.in_features, 200)
        )

    def forward(self, x):
        return self.base(x)

model = CustomResNet50().to(device)

# Optimizer and Scheduler
criterion = nn.CrossEntropyLoss(label_smoothing=0.05)
optimizer = optim.Adam(filter(lambda p: p.requires_grad, model.parameters()), lr=1e-3, weight_decay=1e-4)
scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=30)

# Training Loop
epochs = 30
train_losses, val_losses, train_accuracies, val_accuracies, top5_accuracies = [], [], [], [], []
best_model_wts = copy.deepcopy(model.state_dict())
best_acc = 0.0

for epoch in range(epochs):
    model.train()
    train_loss, correct, total = 0, 0, 0

    for inputs, labels in tqdm(train_loader, desc=f"Epoch {epoch+1} Training"):
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        train_loss += loss.item()
        _, preds = torch.max(outputs, 1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)

    train_acc = correct / total
    train_losses.append(train_loss / len(train_loader))
    train_accuracies.append(train_acc)

    model.eval()
    val_loss, correct, total = 0, 0, 0
    all_preds, all_labels, all_probs = [], [], []
    with torch.no_grad():
        for inputs, labels in tqdm(val_loader, desc=f"Epoch {epoch+1} Validation"):
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            val_loss += loss.item()
            probs = nn.functional.softmax(outputs, dim=1)
            _, preds = torch.max(probs, 1)
            correct += (preds == labels).sum().item()
            total += labels.size(0)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
            all_probs.extend(probs.cpu().numpy())

    val_acc = correct / total
    val_losses.append(val_loss / len(val_loader))
    val_accuracies.append(val_acc)
    top5_acc = top_k_accuracy_score(all_labels, all_probs, k=5)
    top5_accuracies.append(top5_acc)

    print(f"Epoch {epoch+1}/{epochs} - Train Acc: {train_acc:.4f}, Val Acc: {val_acc:.4f}, Top-5: {top5_acc:.4f}")

    if val_acc > best_acc:
        best_acc = val_acc
        best_model_wts = copy.deepcopy(model.state_dict())
        torch.save(best_model_wts, "best_model.pth")

    scheduler.step()

# Load best model
model.load_state_dict(torch.load("best_model.pth"))
model.eval()

# Final Evaluation & Visualization
all_preds, all_labels, all_probs = [], [], []
misclassified_imgs, mis_labels, mis_preds = [], [], []
correct_imgs, correct_labels = [], []

with torch.no_grad():
    for inputs, labels in val_loader:
        inputs, labels = inputs.to(device), labels.to(device)
        outputs = model(inputs)
        probs = nn.functional.softmax(outputs, dim=1)
        _, preds = torch.max(probs, 1)

        for i in range(inputs.size(0)):
            if preds[i] != labels[i] and len(misclassified_imgs) < 8:
                misclassified_imgs.append(inputs[i].cpu())
                mis_labels.append(labels[i].item())
                mis_preds.append(preds[i].item())
            elif preds[i] == labels[i] and len(correct_imgs) < 8:
                correct_imgs.append(inputs[i].cpu())
                correct_labels.append(labels[i].item())

        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())
        all_probs.extend(probs.cpu().numpy())

# Plots
plt.figure()
plt.plot(train_accuracies, label='Train Accuracy')
plt.plot(val_accuracies, label='Validation Accuracy')
plt.plot(top5_accuracies, label='Top-5 Accuracy')
plt.title("Accuracy over Epochs")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.show()

plt.figure()
plt.plot(train_losses, label='Train Loss')
plt.plot(val_losses, label='Validation Loss')
plt.title("Loss over Epochs")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()

# Confusion Matrix (First 10 classes)
cm = confusion_matrix(all_labels, all_preds)
plt.figure(figsize=(10, 8))
sns.heatmap(cm[:10, :10], annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix (Top 10 Classes)")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()

# Class-wise Accuracy Chart
report = classification_report(all_labels, all_preds, output_dict=True)
accuracies = [report[str(i)]['recall'] for i in range(10)]
plt.figure(figsize=(10, 4))
sns.barplot(x=class_names[:10], y=accuracies)
plt.title("Class-wise Accuracy (Top 10 Classes)")
plt.xticks(rotation=90)
plt.ylabel("Accuracy")
plt.show()

# Misclassified Images
fig, ax = plt.subplots(figsize=(12, 6))
ax.set_title("Misclassified Images")
img_grid = make_grid(misclassified_imgs, nrow=8, normalize=True)
ax.imshow(np.transpose(img_grid.numpy(), (1, 2, 0)))
ax.axis('off')
print("True:", [class_names[i] for i in mis_labels])
print("Pred:", [class_names[i] for i in mis_preds])
plt.show()

# Correct Predictions
fig, ax = plt.subplots(figsize=(12, 6))
ax.set_title("Correctly Classified Images")
img_grid = make_grid(correct_imgs, nrow=8, normalize=True)
ax.imshow(np.transpose(img_grid.numpy(), (1, 2, 0)))
ax.axis('off')
print("Labels:", [class_names[i] for i in correct_labels])
plt.show()
```
